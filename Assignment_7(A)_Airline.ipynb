```python
from sklearn.cluster import AgglomerativeClustering
hclusters=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(df3),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
df3['clustersid']=hclusters.labels_
df3
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(df3['clustersid'],df3['Balance'], c=hclusters.labels_) 
```

## K- Mean clustering

## Find the number of K: Elbow method

```python
from sklearn.cluster import KMeans
```

```python
    wcss = []
    for i in range(1,11):
        model= KMeans(n_clusters=i,random_state = 2)
        model.fit(df1)
        a = model.inertia_
        wcss.append(a)


    wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel("Number of K")
plt.ylabel("WCSS")
plt.grid()
```

```python
kmodel= KMeans(n_clusters=5) # by the Elbow method we select n-cluster as 5
```
